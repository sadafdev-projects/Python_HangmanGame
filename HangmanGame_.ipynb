```python
"""
ğŸ¯ Hangman Game (Improved Version)
=================================
A simple text-based Hangman game implemented in Python.

ğŸ“˜ Description:
The player tries to guess a randomly chosen word one letter at a time.
Each incorrect guess reduces the number of remaining attempts (6 total).
The game ends when the player either guesses all letters correctly or runs out of attempts.

ğŸ§  Key Concepts Used:
- random (to select a random word)
- while loop (for repeated guessing until win/loss)
- if-else (for decision making)
- strings and lists (for storing and displaying letters)

âš™ï¸ Simplified Scope:
- 5 predefined words (no file or API)
- 6 incorrect guesses allowed
- Basic console input/output (no graphics or audio)
- Fully responsive in Google Colab

Author: [Your Name]
Date: [Insert Date]
"""

import random

def hangman():
    """Main function to run the Hangman game."""

    # Step 1: Predefined list of 5 words
    words = ["python", "google", "colab", "hangman", "coding"]

    # Step 2: Choose a random word
    secret_word = random.choice(words)

    # Step 3: Initialize variables
    guessed_letters = []   # Stores all guessed letters
    attempts_left = 6       # Player has 6 incorrect guesses

    print("\nğŸ® Welcome to Hangman!")
    print("Try to guess the secret word one letter at a time.")
    print("You have 6 incorrect guesses allowed.\n")

    # Step 4: Main game loop
    while attempts_left > 0:
        # Build the current display of the word
        display_word = ' '.join([letter if letter in guessed_letters else '_' for letter in secret_word])

        print("=" * 35)
        print(f"Word: {display_word}")
        print(f"Guessed Letters: {' '.join(guessed_letters) if guessed_letters else 'None'}")
        print(f"Attempts Left: {attempts_left}")
        print("=" * 35)

        # Step 5: Player input
        guess = input("ğŸ‘‰ Enter a letter: ").lower().strip()

        # Step 6: Validate input
        if len(guess) != 1 or not guess.isalpha():
            print("âš ï¸ Please enter only one alphabet letter.\n")
            continue

        if guess in guessed_letters:
            print("âš ï¸ You already guessed that letter. Try another.\n")
            continue

        # Step 7: Record the guess
        guessed_letters.append(guess)

        # Step 8: Check if guess is correct
        if guess in secret_word:
            print("âœ… Correct guess!\n")
        else:
            print("âŒ Wrong guess!\n")
            attempts_left -= 1

        # Step 9: Check if player has guessed the full word
        if all(letter in guessed_letters for letter in secret_word):
            print(f"ğŸ‰ Congratulations! You guessed the word: {secret_word.upper()}")
            break

    else:
        # Step 10: If loop exits normally (player loses)
        print(f"ğŸ’€ Game Over! The correct word was: {secret_word.upper()}")

# Step 11: Run the game
if __name__ == "__main__":
    hangman()
```
